```python
import pandas as pd
```

```python
car_sales=pd.read_csv("car-sales.csv")
```

```python
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str.replace('[\$\,\.]','')
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sales date"]=pd.date_range("1/1/2022",periods=len(car_sales))
car_sales
```

```python
car_sales["Total sales"]=car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(title="Car Sales",x="Sales date",y="Total sales")
```

```python
car_sales.plot(x="Odometer (KM)",y="Total sales",kind="scatter")
```

```python
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease["age"].plot.hist()
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
over_50.plot(kind='scatter',x='age',y='chol',c='oldpeak');
```

```python
##using oo mixed with pyplot
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age',y='chol',c='oldpeak',ax=ax)
ax.set_xlim([50,100])
```

```python
##oo from strach
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50['age'], ##this is the extraction of all the elements from over 50 and plot it into a graph scatter
          y=over_50['chol'],
            c=over_50['target']);
ax.set(title='age and cholesterol levels',xlabel='age',ylabel='cholesterol levels'); ##set axes hehe
ax.legend(*scatter.legend_elements(),title='target'); ##get legend
ax.axhline(over_50['chol'].mean(),linestyle='--'); ##dashed line
```

```python
plt.style.use('seaborn-whitegrid');
fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
##data for axis 1
scatter=ax1.scatter(x=over_50['age'], ##this is the extraction of all the elements from over 50 and plot it into a graph scatter
          y=over_50['chol'],
            c=over_50['target'],
                   cmap="winter");
ax1.set(title='age and cholesterol levels',ylabel='cholesterol levels'); ##set axes hehe
ax1.set_xlim([50,80])
ax1.legend(*scatter.legend_elements(),title='target'); ##get legend
ax1.axhline(over_50['chol'].mean(),linestyle='--'); ##dashed line
##data for axis 2
scatter=ax2.scatter(x=over_50['age'], ##this is the extraction of all the elements from over 50 and plot it into a graph scatter
          y=over_50['thalach'],
            c=over_50['target'],
                   cmap='summer');
ax2.set(title='age and max heart rate levels',xlabel='age',ylabel='max heart rate levels'); ##set axes hehe
ax2.legend(*scatter.legend_elements(),title='target'); ##get legend
ax2.axhline(over_50['thalach'].mean(),linestyle='--'); ##dashed line
ax2.set_ylim([80,200]);
fig.suptitle("Heart disease analysis",fontsize=30,fontweight='bold');
```
